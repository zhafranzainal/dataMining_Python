```python
# Import related libraries and load the dataset.
import pandas as pd
import numpy as np

df = pd.read_csv('Advertising.csv')
df
```

```python
# Check and solve the noise in the dataset.
df = df.drop(columns=df.columns[0])
df.isnull().sum()
```

```python
# Set the input attributes and target attributes (Advertising.csv = Sales).
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X
```

```python
y
```

```python
# Partition the datasets into training and testing sets with 80:20.
# Discuss the requirement to do data partitioning in developing the model.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 0)
y_test
```

```python
# Train linear regression algorithm for training sets. Explain the model developed.
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Use the linear regression models to predict test sets.
y_pred = regressor.predict(X_test)
y_pred
```

```python
# Evaluate the model using R^2.
import statsmodels.formula.api as sm
result = sm.ols("Sales~TV+Radio+Newspaper", df).fit()
print(result.rsquared, result.rsquared_adj)
```

```python
# Evaluate the model using Mean Squared Error (MSE) for linear regression.
from sklearn.metrics import mean_squared_error
print('MSE (built-in function) :', round(mean_squared_error(y_test, y_pred), 4))

err = np.matrix(y_test-y_pred)
print('MSE (computed)          : %.4f' % (err*err.T/err.shape[1]))
```
